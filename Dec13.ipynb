```julia
# creating a function for the Couette Flow
u(y,Uc,H) = 4*Uc/H^2*y*(H-y)
```

```julia
H = 0.01
Uc = 1.0
y = range(0,H,length=101)
```

```julia
ps = plot(u.(y,Uc,H), y, xlim=(0,2Uc), ylim=(0,H),
    legend=false, xlabel="u(y)", ylabel="y (m)")
```

```julia
ps = plot(u.(y,Uc,H)/Uc, y/H, xlim=(0,2), ylim=(0,1),
    legend=false, xlabel="u(y)/Uc", ylabel="y/H (m)")
```
